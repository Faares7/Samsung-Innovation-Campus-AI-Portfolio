```python
X = df.drop(columns=['price_range'])
y = df['price_range']

X.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score




models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)
    results.append([name, train_acc, test_acc])

results_df = pd.DataFrame(results, columns=["Model", "Train Accuracy", "Test Accuracy"])
results_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(results_df["Model"], results_df["Train Accuracy"], marker='o', label="Train Accuracy")
plt.plot(results_df["Model"], results_df["Test Accuracy"], marker='s', label="Test Accuracy")
plt.xticks(rotation=45)
plt.ylabel("Accuracy")
plt.title("Model Performance Comparison")
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

X = df.drop(columns=['price_range'])
y = df['price_range']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kf = KFold(n_splits=5, shuffle=True, random_state=42)

models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

results = []
for name, model in models.items():
    y_pred = cross_val_predict(model, X_scaled, y, cv=kf)
    train_scores = cross_val_score(model, X_scaled, y, cv=kf, scoring='accuracy')
    test_scores = accuracy_score(y, y_pred)
    results.append([name, train_scores.mean(), train_scores.std(), test_scores])

results_df = pd.DataFrame(results, columns=["Model", "Mean Train Accuracy", "Std Dev Train", "Test Accuracy"])

plt.figure(figsize=(10, 5))
x = range(len(results_df))
plt.bar(x, results_df["Mean Train Accuracy"], width=0.4, label="Train Accuracy", color='blue', align='center')
plt.bar(x, results_df["Test Accuracy"], width=0.4, label="Test Accuracy", color='red', align='edge')
plt.xticks(x, results_df["Model"], rotation=45)
plt.ylabel("Accuracy")
plt.title("Train vs Test Accuracy for Classification Models")
plt.ylim(0, 1)
plt.legend()
plt.grid()
plt.show()

results_df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(results_df["Model"], results_df["Mean Train Accuracy"], marker='o', label="Mean Train Accuracy")
plt.plot(results_df["Model"], results_df["Test Accuracy"], marker='s', label="Test Accuracy")
plt.xticks(rotation=45)
plt.ylabel("Accuracy")
plt.title("Model Performance Comparison")
plt.legend()
plt.grid()
plt.show()
```

```python
from hyperopt import fmin, tpe, hp, Trials


def objective(params):
    model_type = params['model_type']
    del params['model_type']

    if model_type == "Random Forest":
        model = RandomForestClassifier(**params, random_state=42)
    elif model_type == "Gradient Boosting":
        model = GradientBoostingClassifier(**params, random_state=42)
    elif model_type == "SVM":
        model = SVC(**params)

    model.fit(X_scaled, y)
    accuracy = accuracy_score(y, model.predict(X_scaled))
    return -accuracy

space = hp.choice('classifier_type', [
    {'model_type': 'Random Forest', 'n_estimators': hp.choice('n_estimators', [50, 100, 200]), 'max_depth': hp.choice('max_depth', [None, 10, 20]), 'min_samples_split': hp.choice('min_samples_split', [2, 5, 10])},
    {'model_type': 'Gradient Boosting', 'n_estimators': hp.choice('gb_n_estimators', [50, 100, 200]), 'learning_rate': hp.choice('learning_rate', [0.01, 0.1, 0.2]), 'max_depth': hp.choice('gb_max_depth', [3, 5, 10])},
    {'model_type': 'SVM', 'C': hp.choice('svm_C', [0.1, 1, 10]), 'kernel': hp.choice('kernel', ['linear', 'rbf'])}
])

trials = Trials()
best_params = fmin(fn=objective, space=space, algo=tpe.suggest, max_evals=20, trials=trials)

best_model_type = best_params['classifier_type']
best_accuracy = -min(trials.losses())

best_params, best_model_type, best_accuracy
```
