```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Udemy.csv')
df.head()
```

```python
df.info()
```

# Filling Missing Values

```python
df['price'].unique()
```

```python
df['price'] = df['price'].replace('TRUE', 0)
df['price'] = df['price'].replace('Free', 0)
df['price'] = pd.to_numeric(df['price'])
```

```python
print(df['price'].mean())
print(df['num_reviews'].mean())
print(df['num_subscribers'].mean())
print(df['num_lectures'].mean())
```

```python
print(df['price'].describe())
print(df['num_reviews'].describe())
print(df['num_subscribers'].describe())
print(df['num_lectures'].describe())
```

```python
df['price'] = df['price'].fillna(df['price'].mean())
df['num_reviews'] = df['num_reviews'].fillna(df['num_reviews'].mean())
df['num_subscribers'] = df['num_subscribers'].fillna(df['num_subscribers'].mean())
df['num_lectures'] = df['num_lectures'].fillna(df['num_lectures'].mean())
df['content_duration'] = df['content_duration'].dropna()
df.info()
```

```python
df.head()
```

# Encoding

```python
print(df['is_paid'].nunique())
print(df['level'].nunique())
print(df['subject'].nunique())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['is_paid'] = le.fit_transform(df['is_paid'])
df['level'] = le.fit_transform(df['level'])
df['subject'] = le.fit_transform(df['subject'])
```

```python
df.head()
```

```python
df = df.drop(columns = ['course_title','url'])
df.head()
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['price','num_subscribers','num_reviews','num_lectures']] = scaler.fit_transform(df[['price','num_subscribers','num_reviews','num_lectures']])
df.head()
```

# Feature Engineering


```python
df['content_duration'].unique()
```

```python
df[['date_part', 'time_part']] = df['published_timestamp'].str.split('T', expand=True)

df[['year', 'month', 'day']] = df['date_part'].str.split('-', expand=True)
df[['hour', 'minute', 'second']] = df['time_part'].str.split(':', expand=True)

df = df.drop(columns=['date_part', 'time_part', 'minute', 'second'])

df[['year', 'month', 'day', 'hour']] = df[['year', 'month', 'day', 'hour']].astype(int)
df.head()
```
