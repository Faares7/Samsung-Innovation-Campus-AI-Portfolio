```python
import numpy as np

A = np.array([[3, 4, 5],
              [9, 7, 2],
              [8, 3, 9]])

B = np.array([[2, 1, 4],
              [9, 8, 3],
              [5, 6, 7]])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

BA = B @ A
print("\nB*A:\n", BA)

expr1 = A + BA + BA + B
print("\nA + BA + BA + B:\n", expr1)

expr2 = A - BA - BA - B
print("\nA - BA - BA - B:\n", expr2)

print("\nTranspose of A (A^T):\n", A.T)

det_A = np.linalg.det(A)
print("\nDeterminant of A:", det_A)

if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("\nInverse of A:\n", inv_A)
else:
    print("\nMatrix A is singular, no inverse.")

v = np.array([1, 2, 3])
Av = A @ v
print("\nA x v:\n", Av)

Marks = np.array([[80, 90, 70],
                  [60, 75, 85],
                  [95, 88, 92]])

print("\nMarks Matrix (rows=students, cols=subjects):\n", Marks)

print("\nMarks Transposed (rows=subjects, cols=students):\n", Marks.T)

avg_students = np.mean(Marks, axis=1)
print("\nAverage score for each student:\n", avg_students)

avg_subjects = np.mean(Marks, axis=0)
print("\nAverage score for each subject:\n", avg_subjects)
```
