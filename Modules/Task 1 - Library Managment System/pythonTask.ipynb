```python
class Book():
    def __init__(self, id, name, field, num_papers, writer, borrowed_price, is_borrowed = False, rate = 0):
        self.id = id
        self.name = name
        self.field = field
        self.num_papers = num_papers
        self.writer = writer
        self.is_borrowed = is_borrowed
        self.borrowed_price = borrowed_price
        self.rate = rate
        self.list_reviews = []


    def update(self):
      choice = int(input('''
1. Update ID
2. Update name
3. Update field
4. Update number of papers
5. Update writer
6. Update is borrowed
7. Update borrowed price
8. Update rate
'''))
      match choice:
        case 1:
          self.id= int(input("Enter a new ID of the book: "))
          print(f'the id of the book was updated to {self.id}')
        case 2:
          self.name= input("Enter a new name for the book: ")
          print(f'the name of the book with id {self.id} was updated to {self.name}')
        case 3:
          self.field = input("Enter a new field for the book: ")
          print(f'the field of the book with id {self.id} was updated to {self.field}')
        case 4:
          self.num_papers = int(input("Enter a new number of papers of the book: "))
          print(f'the number of papers of the book with id {self.id} was updated to {self.num_papers}')
        case 5:
          self.writer = input("Enter a new writer of the book: ")
          print(f'the writer of the book with id {self.id} was updated to {self.writer}')
        case 6:
          self.is_borrowed = input("Update whether the book is borrowed or not: (yes | no)")
          if self.is_borrowed.lower() == "yes":
            self.is_borrowed = True
            print(f'the borrow state of the book with id {self.id} was updated to {self.is_borrowed}')
          elif self.is_borrowed.lower() == "no":
            self.is_borrowed = False
            print(f'the borrow state of the book with id {self.id} was updated to {self.is_borrowed}')
        case 7:
          self.borrowed_price = int(input("Enter a new borrow price for the book: "))
          print(f'the borrow price of the book with id {self.id} was updated to {self.borrowed_price}')
        case 8:
          self.rate = float(input("Enter a new rate for the book: "))
          print(f'the rate of the book with id {self.id} was updated to {self.rate}')

    def show_info(self):
      print(f'''ID: {self.id}
Name: {self.name}
Field: {self.field}
Number of papers: {self.num_papers}
Writer: {self.writer}
Is borrowed: {self.is_borrowed}
Borrowed price: {self.borrowed_price}
Rate: {self.rate}''')


    def book_age(self,date_founded):
        age_of_book = 2025 - date_founded
        return age_of_book


    def review(self,review,cat):
        self.list_reviews.append((review,cat))
        return self.list_reviews

    def show_review(self):
        for review,cat in self.list_reviews:
          print(f"Review: {review}\n",
f"Type: {cat}\n")
```

```python
book_dic = {}
def interface():
  while True:
    print('''
1. Add new book
2. Search about book info
3. Add a review  for any book
4. Show book reviews
5. Update book info
6. Exit''')
    choice = int(input('Enter your choice: '))
    match  choice:
      case 1:
        id = int(input("Enter the ID: "))
        name = input("Enter the name: ")
        field = input("Enter the field: ")
        num_papers = int(input("Enter the number of papers: "))
        writer = input("Enter the writer: ")
        is_borrowed = input("Enter wether the book is borrowed or not: (yes | no) ")
        if is_borrowed.lower() == "yes":
            is_borrowed = True
        elif is_borrowed.lower() == "no":
            is_borrowed = False
        borrowed_price = int(input("Enter the borrow price: "))
        rate = float(input("Enter the rate: "))
        b = Book(id, name, field, num_papers, writer, is_borrowed, borrowed_price, rate)
        book_dic[b.id] = b
      case 2:
        id = int(input("Enter the book ID you want to know info about: "))
        book_dic[id].show_info()
      case 3:
        id = int(input("Enter the book ID you want to add a review about: "))
        review = input("Enter your review: ")
        cat = input("Enter the type of review: ")
        book_dic[id].review(review,cat)
      case 4:
        id = int(input("Enter the book ID you want to show its reviews: "))
        book_dic[id].show_review()
      case 5:
        id = int(input("Enter the book ID you want to update its info: "))
        book_dic[id].update()
      case 6:
        print("Thank you for using our program")
        break
  return
```
